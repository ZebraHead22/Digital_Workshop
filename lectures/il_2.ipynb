## Полиморфизм и абстрактные методы

```python
class Geom():
     def get_pr(self):
         raise NotImplementedError ("-1")


class Rectangle(Geom):
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def get_pr(self):
        return 2*(self.w + self.h)


class Square(Geom):
    def __init__(self, a):
        self.a = a

    # def get_pr(self):
    #     return 4*self.a
```

```python
r1 = Rectangle(1, 2)
r2 = Rectangle(3, 4)
s1 = Square(10)
s2 = Square(20)
```

```python
geom = [r1, r2, s1, s2] # А можно сразу писать в список
for g in geom:
    print(g.get_pr())
```

А если нет метода??
1. Можно все классы унаследовать от одного.
2. Добавить генерацию исключения

```python
class Products():
   
    
    def __init__(self, name, price, weight):
        super().__init__()
        print("Вызов init Products")
        self.name = name
        self.price = price
        self.weight = weight

    def get_parameters(self):
        print(f"{self.name}, {self.price}, {self.weight}")


class MixinsLog():
    ID = 0

    def __init__(self):
        print("Вызов init MixinLogs")
        MixinsLog.ID += 1
        self.id = MixinsLog.ID
        

    def show_id(self):
        print(f"{self.id}")


class Vegetables(Products, MixinsLog):
    pass

prod1 = Vegetables("tomat", 30, 1)
prod1.get_parameters()
prod1.show_id()
print(prod1.__dict__)
```
