### Метод \_\_repr__

```python
from typing import Any


class Person:
    
    def __init__(self, name, job=None, pay=0):
        self.name = name
        self.job = job
        self.pay = pay
    
    def lastName (self):
        return self.name.split()[-1]
    
    def giveRaise(self, percent):
        self.pay = int(self.pay * (1 + percent))
    
    def __repr__(self):
        return '[Person]: %s, %s' % (self.name, self.pay)

bob = Person('Bob Smith')
sue = Person('Sue Jones', job='dev', pay=100000)
print(bob.__dict__)
print(sue.__dict__)
print(bob.lastName(), sue.lastName())
sue.giveRaise(.10)
print(sue.__dict__)
print(sue.__repr__())
```

### Создание подклассов

```python
from typing import Any


class Person:
    
    def __init__(self, name, job=None, pay=0):
        self.name = name
        self.job = job
        self.pay = pay
    
    def lastName (self):
        return self.name.split()[-1]
    
    def giveRaise(self, percent):
        self.pay = int(self.pay * (1 + percent))
    
    def __repr__(self):
        return '[Person]: %s, %s, %s' % (self.name, self.pay, self.job)

class Manager(Person):

    def __init__(self, name, pay=0):
        Person.__init__(self, name, 'manager', pay)

    def giveRaise(self, percent, bonus=.10):
        Person.giveRaise(self, percent + bonus)

bob = Manager('Bob Smith', 100000)
bob.giveRaise(0.05)
print(bob.job)
```

```python
tom = Manager('Tom Bark', 50000)
tom.giveRaise(.09)
sue = Manager('Sue Jones', 12000)
print(tom)
print(sue)
```

```python
for obj in (bob, tom, sue):
    obj.giveRaise(0.5)
    print(obj)
```

```python
class Product(Manager):
    def giveRaise(self, percent, bonus=0.3):
        return Person.giveRaise(percent, bonus)
    
    def firstName(self):
        return self.name.split()[0]
```

```python
mark = Product("Mark Sina", 'pro', 90000)
mark.firstName()
bob.firstName()
```

```python
class Developer(): # Наследование через внедрение
    def __init__(self, name, pay):
        self.person = Person(name, 'developer', pay)
    
    def giveRaise(self, percent, bonus = 0.01):
        self.person.giveRaise(percent + bonus)

    def __getattr__(self, attr):
        return getattr(self.person, attr)
    
    def __repr__(self):
        return str(self.person)
    
lis = Developer('Liza Edelt', 200000)
print(lis)
```
